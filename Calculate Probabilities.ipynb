```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv("weight-height - weight-height.csv")
females = df.loc[ df['Gender'] == 'Female'].copy()
females.head()
```

```python
to_plot=females['Height']
```

calculate the probability that a female's height is:

1. between 55.0 and 56.0 inches = 0.015%

2. is less than 5 feet (60 in) = 0.08%

3. is greater than the mean = 50%

```python
low = 55
high = 56
```

```python
mean = np.mean(to_plot)
std = np.std(to_plot)
min = np.min(to_plot)
max = np.max(to_plot)
```

```python
#to visualize
x_range = np.linspace(min, max,1000)

cdf = stats.norm.cdf(x_range, loc=mean, scale=std)

fig, ax = plt.subplots(figsize=(12,6))
sns.histplot(x=to_plot, stat='density', cumulative = True)
ax.plot(x_range,cdf)
ax.axvline(low, color= 'black')
ax.axvline(high, color ='black')
ax.axvspan(low, high, color ='yellow', zorder = 0)
ax.set_title('Range of Interest');
```

```python
#calculating
stats.norm.cdf(high, loc=mean,scale=std) - stats.norm.cdf(low,loc=mean, scale=std)
```

```python
stats.norm.cdf(60, loc=mean,scale=std)
```
